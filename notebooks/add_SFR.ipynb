```python
from astropy.table import Table
import numpy as np
```

```python
FOLDER = '/Users/nityaravi/Documents/Research/RotationCurves/data/manga/output_files/DR17/CURRENT_MASTER_TABLE/'
working_table_fn = FOLDER + 'H_alpha_HIvel_BB_extinction_H2_MxCG_R90_v3p5_Z_SFR.fits'
Pipe3D_fn = '/Users/nityaravi/Documents/Research/RotationCurves/data/manga/DR17/SDSS17Pipe3D_v3_1_1.fits'
```

```python
working_table = Table.read(working_table_fn, format='fits')
Pipe3D_table = Table.read(Pipe3D_fn, format='fits')
```

```python
Pipe3D_dict = {}
for i in range(0, len(Pipe3D_table)):
    key = Pipe3D_table['plateifu'][i]
    Pipe3D_dict[key] = i
    

    
```

```python
working_table['SFR'] = np.ones(len(working_table))*np.nan
working_table['sSFR'] = np.ones(len(working_table))*np.nan
working_table['SFR_err'] = np.ones(len(working_table))*np.nan
working_table['sSFR_err'] = np.ones(len(working_table))*np.nan
```

```python
for i in range(0, len(working_table)):
    gal_ID = working_table['plateifu'][i]
    
    try:
        i_P3 = Pipe3D_dict[gal_ID]
    except:
        continue
    

    working_table['SFR'][i] = Pipe3D_table['log_SFR_Ha'][i_P3]
    working_table['SFR_err'][i] = Pipe3D_table['e_log_SFR_Ha'][i_P3]
    
    if working_table['M_disk'][i] > 0:
        
        SFR = 10**Pipe3D_table['log_SFR_Ha'][i_P3]
        SFR_err = 10**Pipe3D_table['e_log_SFR_Ha'][i_P3]
    
        M_disk = 10**working_table['M_disk'][i]
        M_disk_err = 10**working_table['M_disk_err'][i]
    
        working_table['sSFR'][i] = np.log10(SFR/M_disk)
    
        working_table['sSFR_err'][i] = np.log10(np.sqrt((SFR_err**2/M_disk**2 + SFR**2*M_disk_err**2/M_disk**2)))
    
        
    
```
