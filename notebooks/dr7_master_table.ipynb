# Calculate H2 mass with parametrization

```python
dr7['param_H2'] = np.ones(len(dr7))*-999
dr7['param_H2_err'] = np.ones(len(dr7))*-999
```

```python
def H2_param(Mr, CMD):
    
    if CMD == 1:
        a = -0.4
        a_err = 0.02
        b = 1.12
        b_err = 0.36
    
    elif CMD == 2 or CMD == 3:
        a = -0.27
        a_err = 0.02
        b = 3.62
        b_err = 0.37
    
    else:
        return -999, -999
    
    H2 = a*Mr + b
    H2_err = np.sqrt(a_err**2 * Mr**2 + b_err**2)
    
    return H2, H2_err
```

```python
for i in range(len(dr7)):
    if dr7['DR7_rabsmag_NSA'][i] != 0 and dr7['DR7_CMD_class'][i] > 0:
        
        dr7['param_H2'][i], dr7['param_H2_err'][i] = H2_param(dr7['DR7_rabsmag_NSA'][i],dr7['DR7_CMD_class'][i])
```

# R90 [kpc]

```python
dr7['R90_kpc'] = np.ones(len(dr7))*-999
H_0 = 100      # Hubble's Constant in units of h km/s/Mpc
c = 299792.458 # Speed of light in units of km/s

for i in range(len(dr7)):
    if dr7['nsa_elpetro_th90_r'][i] > 0 and dr7['nsa_z'][i] > 0:
        
        dist_to_galaxy_Mpc = c*dr7['nsa_z'][i]/H_0
        dist_to_galaxy_kpc = dist_to_galaxy_Mpc*1000

        r_kpc = dist_to_galaxy_kpc*np.tan(dr7['nsa_elpetro_th90_r'][i]*(1./60)*(1./60)*(np.pi/180))
        dr7['R90_kpc'][i] = r_kpc
```

# Calculate Metallicity

```python
from metallicity_map import calculate_global_metallicity
dr7['Z12logOH_err'] = np.ones(len(dr7))*-999

for i in range(0, len(dr7)):
    if dr7['Flux_OII_3726'][i] > -999:
        
        if dr7['Flux_OII_3726'][i] ==0 and dr7['Flux_OII_3728'][i]==0:
            continue
        if dr7['Flux_NII_6547'][i]==0 or dr7['Flux_NII_6583'][i]==0 or\
            dr7['Flux_OIII_4958'][i] ==0 or dr7['Flux_OIII_5006'][i]==0 or\
            dr7['Flux_Ha_6562'][i] == 0 or dr7['Flux_Hb_4861'][i] == 0:
                continue
        
        fluxes = {'OII': dr7['Flux_OII_3726'][i],
                    'OII_err': dr7['Flux_OII_3726_Err'][i],
                    'OII2': dr7['Flux_OII_3728'][i],
                    'OII2_err': dr7['Flux_OII_3728_Err'][i], 
                    'NII': dr7['Flux_NII_6547'][i],
                    'NII_err': dr7['Flux_NII_6547_Err'][i],
                    'NII2': dr7['Flux_NII_6583'][i],
                    'NII2_err': dr7['Flux_NII_6583_Err'][i],
                    'OIII': dr7['Flux_OIII_4958'][i],
                    'OIII_err': dr7['Flux_OIII_4958_Err'][i],
                    'OIII2': dr7['Flux_OIII_5006'][i],
                    'OIII2_err': dr7['Flux_OIII_5006_Err'][i],
                    'Ha' : dr7['Flux_Ha_6562'][i],
                    'Ha_err': dr7['Flux_Ha_6562_Err'][i],
                    'Hb': dr7['Flux_Hb_4861'][i],
                    'Hb_err': dr7['Flux_Hb_4861_Err'][i]
                    }
    
        Z, Z_err = calculate_global_metallicity(fluxes)
        
        dr7['Z12logOH'][i], dr7['Z12logOH_err'][i] = Z, Z_err
```

# Calculate dust mass

```python
from metallicity_map import calculate_metal_mass

dr7['Mdust_err']=np.ones(len(dr7))*-999

for i in range(len(dr7)):
    if dr7['Z12logOH'][i] > 0 and dr7['logHI'][i] > 0 and dr7['param_H2'][i] > 0:
            
            
            Z = dr7['Z12logOH'][i]
            Z_err = dr7['Z12logOH_err'][i]
            M_HI = 10**dr7['logHI'][i]
            M_HI_err = 0
            M_H2 = 10**dr7['param_H2'][i]
            M_H2_err = 10**dr7['param_H2_err'][i]


            
            Mztot, Mztot_err = calculate_metal_mass(Z, Z_err, M_HI, M_HI_err, M_H2, M_H2_err, None, None)
            print('Mztot ', Mztot, ' Mztot_err ', Mztot_err)

            dr7['Mdust'][i] = Mztot
            dr7['Mdust_err'][i] = Mztot_err
```
