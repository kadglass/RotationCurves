## Methods to extract masses

```python
def get_HI_ALFALFA_mass(gal_id):
    '''
    Find galaxy in HI_MaNGA catalog and extract HI mass and upper limit of HI mass
    
    PARAMETERS
    ==========
    
    gal_id : string
        [PLATE-IFU]
    
    RETURNS
    =======
    
    MHI : float
        HI mass in log scale
    MHI_lim
        upper limit of HI mass in log scale
    
    
    
    '''
    
    index = -1
    print(gal_id)
    for i in range(0, len(A_table['PLATEIFU'])):
        if A_table['PLATEIFU'][i] == gal_id:
            index = i
    
    if index == -1:
        MHI = 0
        MHI_lim = 0
            
    else:
        MHI = A_table['LOGMHI'][index]
        MHI_lim = A_table['LOGHILIM200KMS'][index]
    
    print(gal_id + '\n ALFALFA MHI: ' + str(MHI) +' MHI_lim: ' + str(MHI_lim))
    return MHI, MHI_lim
    
```

```python
def get_HI_mass(gal_id):
    '''
    Find galaxy in ALFALFA catalog and extract HI mass and upper limit of HI mass
    
    PARAMETERS
    ==========
    
    gal_id : string
        [PLATE-IFU]
    
    RETURNS
    =======
    
    MHI : float
        HI mass in log scale
    MHI_lim
        upper limit of HI mass in log scale
    
    
    
    '''    
    
    index = -1
    for i in range(0,len(HI_data['PLATEIFU'])):
        if HI_data['PLATEIFU'][i] == gal_id:
            index = i
            
    if index == -1:
        MHI = 0
        MHI_lim = 0
    
    else:
        MHI = HI_data['LOGMHI'][index]
        MHI_lim = HI_data['LOGHILIM200KMS'][index]
        
    print(gal_id + '\n MHI: ' + str(MHI) +', MHI_lim: ' + str(MHI_lim))
    return MHI, MHI_lim
```

```python
def get_H2_mass(gal_id):
    '''
    Look for galaxy in MASCOT catalog and extract H2 mass. If not in MASCOT catalog, look for galaxy in xCOLD GASS
    catalog and extract H2 mass
    
    PARAMETERS
    ==========
    
    gal_id : string
        [PLATE-IFU]
    
    RETURNS
    =======
    
    MH2 : float
        H2 mass in log scale
    MH2_err: float
        error in H2 (0 for MASCOT galaxies)
    MH2_lim : float
        upper limit for H2 mass in log scale (0 for MASCOT galaxies)
    
    
    
    
    '''   
        
        
    in_mascot = False
    index = -1
    
    for i in range(0,len(MASCOT_data['MaNGA ID'])):
        if MASCOT_data['MaNGA ID'][i] == gal_id:
            in_mascot = True
            index = i
            

    if in_mascot == True:        
        MH2 = MASCOT_data['log(H2_mass)'][index]
        MH2_err = 0
        MH2_lim = 0
    
    elif in_mascot == False:
        drp_index = 0
        for i in range(0,len(DRP_data['plateifu'])):
            if DRP_data['plateifu'][i] == gal_id:
                drp_index = i
        nsa_iau = DRP_data['nsa_iauname'][drp_index]
        
        for i in range(0,len(xCG_data['SDSS'])):
            if xCG_data['SDSS'][i] == nsa_iau:
                index = i
        MH2 = xCG_data['LOGMH2'][index]
        MH2_err = xCG_data['LOGMH2_ERR'][index]
        MH2_lim = xCG_data['LIM_LOGMH2'][index]
    
    print(gal_id + '\n MH2: ' + str(MH2) +', MH2_err: ' + str(MH2_err) + ', MH2_lim: ' + str(MH2_lim))
    return MH2, MH2_err, MH2_lim
        
```

```python
# construct arrays to hold all the HI and H2 values
num_gal = len(FILE_IDS)

# HI-MaNGA HI masses
MHI = np.zeros(num_gal)
MHI_lim = np.zeros(num_gal)

#ALFALFA HI masses
MHI_A = np.zeros(num_gal)
MHI_A_lim = np.zeros(num_gal)

#H2 masses
MH2 = np.zeros(num_gal)
MH2_err = np.zeros(num_gal)
MH2_lim = np.zeros(num_gal)

```

```python
#fill in the columns!


for i in range(0,len(FILE_IDS)):
    gal_id = FILE_IDS[i]
    MHI[i], MHI_lim[i] = get_HI_mass(gal_id)
    MHI_A[i], MHI_A_lim[i] = get_HI_ALFALFA_mass(gal_id)
    MH2[i], MH2_err[i], MH2_lim[i] = get_H2_mass(gal_id)
```

```python
col1 = fits.Column(name='ID', format = '32A', array=FILE_IDS)
col2 = fits.Column(name='MHI', format = 'D', array=MHI)
col3 = fits.Column(name='MHI_lim', format = 'D', array=MHI_lim)
col4 = fits.Column(name = 'MHI_A', format = 'D', array=MHI_A)
col5 = fits.Column(name = 'MHI_A_lim', format = 'D', array=MHI_A_lim)
col6 = fits.Column(name='MH2', format = 'D', array=MH2)
col7 = fits.Column(name='MH2_err', format = 'D', array=MH2_err)
col8 = fits.Column(name='MH2_lim', format = 'D', array=MH2_lim)
```

```python
coldefs = fits.ColDefs([col1, col2, col3, col4, col5, col6, col7, col8])
```
