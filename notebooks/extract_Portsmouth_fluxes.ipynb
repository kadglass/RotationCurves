```python
import numpy as np
from astropy.table import Table
```

```python
DATA_FOLDER = '/Users/nityaravi/Documents/Research/RotationCurves/data'
MASTER_FOLDER = DATA_FOLDER + '/manga/output_files/DR17/CURRENT_MASTER_TABLE/'
portsmouth_fn = DATA_FOLDER + '/NSA_v1_0_1_vflag_Portsmouth.fits'
working_fn = MASTER_FOLDER + 'H_alpha_HIvel_BB_extinction_H2_MxCG_R90_v3p5_Z_SFR.fits'
```

```python
portsmouth = Table.read(portsmouth_fn, format='fits')
working_table = Table.read(working_fn, format='fits')
```

```python
portsmouth_dict = {}
for i in range(0,len(portsmouth)):
    key = portsmouth['NSAID'][i]
    portsmouth_dict[key] = i
```

```python
working_table['Flux_OII_3726'] = np.zeros(len(working_table))
working_table['Flux_OII_3726_Err'] = np.zeros(len(working_table))
working_table['Flux_OII_3728'] = np.zeros(len(working_table))
working_table['Flux_OII_3728_Err'] = np.zeros(len(working_table))
working_table['Flux_OIII_4958'] = np.zeros(len(working_table))
working_table['Flux_OIII_4958_Err'] = np.zeros(len(working_table))
working_table['Flux_OIII_5006'] = np.zeros(len(working_table))
working_table['Flux_OIII_5006_Err'] = np.zeros(len(working_table))
working_table['Flux_NII_6547'] = np.zeros(len(working_table))
working_table['Flux_NII_6547_Err'] = np.zeros(len(working_table))
working_table['Flux_NII_6583'] = np.zeros(len(working_table))
working_table['Flux_NII_6583_Err'] = np.zeros(len(working_table))
working_table['Flux_Ha_6562'] = np.zeros(len(working_table))
working_table['Flux_Ha_6562_Err'] = np.zeros(len(working_table))
working_table['Flux_Hb_4861'] = np.zeros(len(working_table))
working_table['Flux_Hb_4861_Err'] = np.zeros(len(working_table))
```
