```python
Ha_file = '/Users/nityaravi/Documents/Research/data/manga/output_files/DRP_HaVel_map_results_BB_smooth_lt_2.0_.fits'
hdul = fits.open(Ha_file)
Ha_table = Table.read(Ha_file, format='fits')
hdul.close()
```

```python
DESI_phot_file = '/Users/nityaravi/Documents/Research/MaNGA_DESI_r_Photometry.csv'
DESI_table = Table.read(DESI_phot_file, format='csv')
DESI_table[0]
```

```python
MANGA_SPAXEL_SIZE = 0.5*(1/60)*(1/60)*(np.pi/180) # spaxel size (0.5") in radians
H_0 = 100      # Hubble's Constant in units of h km/s/Mpc
c = 299792.458 # Speed of light in units of km/s
q0 = 0.2 # nominal disk thickness
```

```python
Z12LOGOH_r25 = np.zeros(len(FILE_IDS))
Z12LOGOH_R50 = np.zeros(len(FILE_IDS))
```

```python
'''
fit the metallicity to a linear model of metallicity vs r
use the model to find the metallicity at the R_25 radius (extracted from DESI)
use the model to find metallicity at the R_50 Petrosian radius





'''


for i in range(0, len(FILE_IDS)):
    
    gal_id = FILE_IDS[i]
    print(gal_id)
    
    met_fn = '/Users/nityaravi/Documents/Research/data/manga/MetallicityFITS_Pilyugin_DR17_v3/'+ gal_id +'_Pilyugin_DR17_v3.fits'
    hdul = fits.open(met_fn)
    met_map = hdul[1].data
    hdul.close()
    
    i_Ha = -1
    for j in range(0, len(Ha_table['plateifu'])):
        if Ha_table['plateifu'][j] == gal_id:
            i_Ha = j
    i_DESI = -1
    for j in range(0, len(DESI_table['PlateIFU'])):
        if DESI_table['PlateIFU'][j] == gal_id:
            i_DESI = j
    
    if i_DESI == -1:
        print('not in DESI')
        
    r25_arcsec = DESI_table['R25'][i_DESI]
    
        
    z = Ha_table['z'][i_Ha]
    print('z: ' + str(z))

    dist_to_galaxy_Mpc = c*z/H_0
    dist_to_galaxy_kpc = dist_to_galaxy_Mpc*1000

    pix_scale_factor = dist_to_galaxy_kpc*np.tan(MANGA_SPAXEL_SIZE) 
    
    r25_kpc = dist_to_galaxy_kpc*np.tan(r25_arcsec*(1./60)*(1./60)*(np.pi/180))
        
    
    center = (Ha_table['x0'][i_Ha],Ha_table['x0'][i_Ha])
    print(center)
    phi = Ha_table['phi'][i_Ha]
    cosi_angle = (Ha_table['ba'][i_Ha]**2 - q0**2)/(1 - q0**2)
    i_angle = np.arccos(np.sqrt(cosi_angle))
    
    
    count = 0
    for j in range(0,len(met_map)):
        for k in range(0,len(met_map)):
            if met_map[j][k]>8:
                count = count + 1

    r = np.zeros(count)
    Z = np.zeros(count)
    r_i = 0
    for j in range(0,len(met_map)):
        for k in range(0, len(met_map)):
            if met_map[j][k] > 8 and met_map[j][k] < 15:
                r[r_i] = pix_scale_factor*deproject_spaxel((j,k), center, phi, i_angle)[0]
                Z[r_i] = met_map[j][k]
                r_i = r_i + 1
                
    fit = np.polyfit(r,Z,1)
    print(fit)
    
    if i_DESI == -1:
        metallicity_r25 = 0
        
        
    else:
        metallicity_r25 = fit[0]*r25_kpc*0.4 + fit[1]
    
    
    metallicity_R50 = fit[0] * Ha_table['nsa_elpetro_th50_r'][i_Ha]*0.4 +fit[1]
    print('metallicity_r25: ' + str(metallicity_r25))
    print('metallicity_R50: ' + str(metallicity_R50))
    
    Z12LOGOH_r25[i] = metallicity_r25
    Z12LOGOH_R50[i] = metallicity_R50
    
```

```python
col1 = fits.Column(name='ID', format = '32A', array=FILE_IDS)
col2 = fits.Column(name='Z12LOGOH_r25', format = 'D', array=Z12LOGOH_r25)
col3 = fits.Column(name='Z12LOGOH_R50', format = 'D', array=Z12LOGOH_R50)
```
